```python
from DataLoader import DataLoader
from LR import LR
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
train_loader = DataLoader('./input/train.txt')
test_loader = DataLoader('./input/test.txt')
```

```python
lr_params = {
    'train_loader': train_loader,
    'test_loader': test_loader,
    'opt_algo': 'gd',
    'learning_rate': 0.1,
    'epoch': 500,
    'early_stop_round': 10
}
```

```python
lr_model = LR(**lr_params)
```

```python
lr_model.run() 
```

```python
fig = plt.figure()
plt.grid()
plt.plot(lr_model.train_scores, label='train')
plt.plot(lr_model.test_scores, label='test')
plt.legend()  
```

```python
fig = plt.figure()
plt.grid()
plt.plot(lr_model.time_scores, lr_model.train_scores, label='train')
plt.plot(lr_model.time_scores, lr_model.test_scores, label='test')
plt.legend()  
```
